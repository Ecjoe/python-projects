```python
#importing libraries

import random
import sys

#global vaues
MAX_VAL = 100
DICE_FACE = 6

#position of snakes and where snake bite leads to
snakes={17 : 7, 54 : 34, 62 : 19, 98 : 79}

#position of ladders and where it leads to
ladders={3 : 38, 24 : 33, 42 : 93, 72 : 84}

#snake bite info
def snake_bite(current_value, final_value , player_name):
    
    print("You have bitten by snake at ",current_value )
    print("Your position is ",final_value )
    
#ladder jumps info    
def ladder_jump(current_value, final_value , player_name):
    
    print("You have got a ladder jump at ",current_value )
    print("Your position is ",final_value )

#game modes for snake-ladder game
def playing_mode(argument): 
    
    switcher= {
        0: "Auto",
        1: "Manual",
    }
    
    return switcher.get(argument, "Try Again")

#generating random dice values
def get_dice_value():
    
    dice_value = random.randint(1,DICE_FACE)
    print("You got a " + str(dice_value))
    return dice_value

#gameplay for snake-ladder game
def game(player_name, player_pos, dice_value):
    
    old_value = player_pos
    current_value = old_value + dice_value
    
    if current_value > MAX_VAL:
        print("Your final postion is ",old_value)
        return old_value
        
    if current_value in snakes:
        final_value = snakes.get(current_value)
        snake_bite(current_value, final_value , player_name)
        
    elif current_value in ladders:
        final_value = ladders.get(current_value)
        ladder_jump(current_value, final_value , player_name)
        
    else:
        final_value = current_value
        
    print("Your final postion is ",final_value)
        
    return final_value

#winning check rule
def check_win(player_name, player_pos):
    
    if player_pos == MAX_VAL:
        
        print("##### " + player_name + " won the game #####")
        print("##### Game successfully finished #####")
        sys.exit()


def main():
    print("##### Welcome to Snakes & Ladders Game #####")
    
    #enter player names
    player1_name = input("Enter the name of Player 1 : ")
    player2_name = input("Enter the name of Player 2 : ")
    
    print("##### Let us Start #####")
    
    #initial position of players
    player1_current_pos = 0
    player2_current_pos = 0
    
    #choosing player modes
    print("Type '0' : Auto\nType '1' : Manual")
    game_type = int(input("type : "))
    print("choice is " + playing_mode(game_type))
    
    flag = 0
    
    #game in auto mode
    if game_type == 0:
        while True:
            if flag%2 == 0:
                
                #enter roll or any input
                chance = input("Player1_value : ")

                if chance == 'roll' :

                    #getting dice values , position of players and checking win of player 1
                    flag=flag+1
                    dice_value = get_dice_value()
                    player1_current_pos = game(player1_name, player1_current_pos, dice_value)
                    check_win(player1_name, player1_current_pos)

                elif chance == 'quit' :

                    #if player 1 wants to quit , player 2 wins
                    print("##### Player 2 wins #####")
                    sys.exit()

                else:
                    
                    #if wrong input by the user
                    print("Wrong Input , Try Again ")
                    continue
                        
            if flag%2 != 0:
                #enter roll or any input
                chance = input("Player2_value : ")
                
                if chance == 'roll' :

                    #getting dice values , position of players and checking win of player 1
                    flag=flag+1
                    dice_value = get_dice_value()
                    player2_current_pos = game(player2_name, player2_current_pos, dice_value)
                    check_win(player2_name, player2_current_pos)

                elif chance == 'quit' :

                    #if player 2 wants to quit , player 1 wins
                    print("##### Player 1 wins #####")
                    sys.exit()

                else:
                    
                    #if wrong input by the user
                    print("Wrong Input , Try Again ")
                    continue
                
                
    #game in manual mode            
    if game_type == 1:
        while True:
            if flag%2 == 0:
                #enter any input
                chance = (input("Player1_value : "))
                if(chance.isdigit()):
                    if int(chance)>= 1 and int(chance)<= 20 :

                        #getting dice values , position of players and checking win of player 1
                        flag=flag+1
                        dice_value = get_dice_value()
                        player1_current_pos = game(player1_name, player1_current_pos, dice_value)
                        check_win(player1_name, player1_current_pos)

                    elif chance == 'quit' :

                        #if player1 wants to quit , player 2 wins
                        print("##### Player 2 wins #####")
                        sys.exit()
                        
                    else:
                        #if wrong input by the user
                        print("Wrong Input , Try Again ")
                        continue

                else:
                    #if wrong input by the user
                    print("Wrong Input , Try Again ")
                    continue
                
            if flag%2 != 0:
                #enter any input
                chance = str(input("Player2_value : "))
                if(chance.isdigit()):
                    
                    if int(chance)>= 1 and int(chance)<= 20 :

                        #getting dice values , position of players and checking win of player 1
                        flag=flag+1
                        dice_value = get_dice_value()
                        player2_current_pos = game(player2_name, player2_current_pos, dice_value)
                        check_win(player2_name, player2_current_pos)
                        
                    elif chance == 'quit' :

                        #if player 2 wants to quit , player 1 wins
                        print("##### Player 1 wins #####")
                        sys.exit()
                        
                    else:
                       #if wrong input by the user
                        print("Wrong Input , Try Again ")
                        continue

                else:
                    
                   #if wrong input by the user
                    print("Wrong Input , Try Again ")
                    continue
                         
if __name__ == '__main__':
    main()
```
